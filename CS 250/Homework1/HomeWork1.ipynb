```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('avocado-updated-2020.csv')

# Display the first few rows of the dataframe to understand its structure
print(df.head())

# Ensure the 'Date' column is in datetime format
df['date'] = pd.to_datetime(df['date'])

# Extract the year from the 'Date' column
df['year'] = df['date'].dt.year

# a) Bar chart with the average yearwise price of avocado

# Group by year and calculate the average price
average_price_per_year = df.groupby('year')['average_price'].mean()

# Plotting
plt.figure(figsize=(10, 6))
average_price_per_year.plot(kind='bar', color='skyblue')
plt.title('Average Yearwise Price of Avocado')
plt.xlabel('year')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

# b) Line chart with the total volume of avocados sold per year

# Group by year and calculate the total volume
total_volume_per_year = df.groupby('year')['total_volume'].sum()

# Plotting
plt.figure(figsize=(10, 7))
total_volume_per_year.plot(kind='line', marker='o', color='green')
plt.title('Total Volume of Avocados Sold per Year')
plt.xlabel('year')
plt.ylabel('Total Volume')
plt.grid(True)
plt.show()
```
